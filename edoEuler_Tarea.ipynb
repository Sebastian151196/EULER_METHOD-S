<center><h3>Método de Euler</h3></center>

El método de Euler aproxima numéricamente las soluciones de ecuaciones diferenciales ordinarias de primer orden (EDO) con un valor inicial dado.

La ecuacion de primer orden debe ser proporcionado de la siguiente manera:

<center>$ \frac{dy}{dt} = f(t, y), a \leq t \leq b, y(a) = a  $</center>

La ecuación diferencial describe cómo se comporta la derivada de $y$ , como una función de una variable independiente $t$, y la función $y(t)$. Queremos determinar si la función $y(t)$ para nuestra condición inicial $y(a) = a$, en el dominio $a \le t \le b$.

El método de Euler divide el dominio en $N$ piezas de tamaño $h$. Si el dominio esta dado por $[a, b]$, entonces el tamaño del paso está dado por $h = \frac{(x_f - x_0)}{N}$. Despues de calcular $h$, el método usa la derivado para aproximar el siguiente valor de la función subyacente. La derivada de la función $y(t)$ está dada por:

<center> $ y´(t) = \frac{y(t + h) - y(t)}{h},$ tal que $, h \rightarrow 0 $ </center>



<center><h4>Calculo del Error</h4></center>

Podemos usar la siguiente expresión que nos evalurara el error absoluto, que es nada mas que la suma del valor absoluto de los residuos. 

Esto es:

\begin{gather}
  E_{abs} = \sum_{i=1}^{N} |{y_{t_{i}} - w_{i}}|
\end{gather}



<center><h4>Calculo del Error Relativo</h4></center>

El error relativo esta dado por:

\begin{gather}
  E_{rel} = \sum_{i=1}^{N} \frac{|w_{i} - y_{t_{i}}|}{|y_{t_{i}}|}
\end{gather}

```python
import numpy as np 
from matplotlib.pyplot import *

print("-----------Metodo de Euler-----------\n")

# Pediremos al usuario que ingrese los limites
print("Ingrese los limites tales que: a <= t <= b\n")
a = float(input("Dame el limite a = "))
b = float(input("\nDame el limite b = "))

#Pediremos que el usuario ingrese el numero de los pasos
N = int(input("\nIngrese el numero de pasos: "))

h = (b-a)/int(N)  

# Supondremos el valor inicial de y(0) = 0.5

y0 = float(input("\nIngrese el valor de y(0): "))

ValInicial = (0.0, y0)

formula = raw_input("\nIngrese la Funcion: y' = ")

# Definiremos una funcion para la ecuacion diferencial 
def f(t, y):
    return eval(formula)
    
# Creamos unos contenedores para la solucion
t = np.arange( a, b+h, h )
w = np.zeros((N+1,))

 # Colocamos los valores iniciales                      
t[0], w[0] = ValInicial 

# Aplicamos el metodo de Euler
for i in range(1,N+1):
    w[i] = w[i-1] + h * f( t[i-1], w[i-1] )
    
    
# Recuperamos la solucion exacta

solExacta = raw_input("\nIngrese la Funcion de Solucion Exacta: y(t) = ")

def y( t ):
    return eval(solExacta)
    # Sol. Exacta: (t+1.0)**2.0-0.5*np.exp(t)        

print("\n-----------------------------------------------------\n")
print("Valor Metodo de Euler: \n")
print w[i]
print("\n-----------------------------------------------------\n")


error_abs = lambda y, w: np.sum( np.abs( y - w ) )
error_abs( y(t), w )

print("\n-----------------------------------------------------\n")
print("El Error Absoluto en el metodo es: \n")
print (error_abs(y(t), w))
print("\n-----------------------------------------------------\n")


error_rel = lambda y, w: np.sum( np.abs( y - w ) / np.abs( y ) )
error_rel( y(t), w )

print("\n-----------------------------------------------------\n")
print("El Error Relativo en el metodo es: \n")
print (error_rel(y(t), w))
print("\n-----------------------------------------------------\n")


plot( t, w, 'y', label='Aproximacion' )
plot( t, y(t), 'r' , label='Valor Exacto' )
title( "Euler's Method, N="+str(N) )
xlabel('t') 
ylabel('y(t)')
legend(loc=4)
grid()
```

<center><h3>Método de Euler Mejorado en el Medio Punto</h3></center>

Este método se obtiene de aplicar a la expresión:

\begin{align}
    x(t_{1}) = x(t_{0}) \int_{t_{0}}^{t_{1}} f(s, x(s)) ds
\end{align}

la fórmula en el punto medio, quedando lo siguiente:

\begin{align}
    x(t_{1}) = x(t_{0}) + (t_{1} - t_{0}) f \cdot (t_{0} + \frac{t_{1} - t_{0} }{2}), x \cdot (\frac{t_{1} - t_{0}}{2}))  
\end{align}

O sea, siendo $x_{\frac{1}{2}} = x \cdot (t_{0} + (\frac{t_{1} - t_{0}}{2})) $, 

\begin{align}
    x_{1} = x_{0} + h \cdot f \cdot (t_{0} + \frac{h}{2}, x_{\frac{1}{2}}).
\end{align}

En general, el método de Euler modificado viene dado por la recurrencia de:

\begin{align}
    x_{k+1} = x_{k} + h \cdot f \cdot (t_{k} + \frac{h}{2}, x_{\frac{k+1}{2}}), k = 0, 1, ..., n-1.
\end{align}

Este método no es implícito, pues no viene dado $x_{k+1}$ en función de sí mismo, pero tampoco es explícito, pues necesitamos determinar previamente $x_{\frac{k+1}{2}}$. Como es en el Caso de Heun, usaremos Euler para hayar el valor previo. Por tanto, no se puede considerar que Euler modificado es un método de dos pasos. 

Así pues el método de Euler Modificado viene dado por la recurrencia:

\begin{align}
    Predicción: x_{\frac{k+1}{2}} = x_{k} + \frac{h}{2} \cdot f \cdot (t_{k}, x_{k})
    \\\
    Correción: x_{k+1} = x_{k} + h \cdot f \cdot (t_{k} + \frac{h}{2}, x_{\frac{k+1}{2}})
\end{align}


```python
import numpy as np 
from matplotlib.pyplot import *

print("-----------Metodo de Euler Mejorado-----------\n")

# Pediremos al usuario que ingrese los limites
print("Ingrese los limites tales que: a <= t <= b\n")
a = float(input("Dame el limite x0 = "))
b = float(input("\nDame el limite xf = "))

#Pediremos que el usuario ingrese el numero de los pasos
N = int(input("\nIngrese el numero de pasos: "))

h = (b-a)/int(N)  

y0 = float(input("\nIngrese el valor de y(0): "))

ValInicial = (0.0, y0)

formula = raw_input("\nIngrese la Funcion: y' = ")

# Definiremos una funcion para la ecuacion diferencial 
def f(t, y):
    return eval(formula)
    
# Creamos unos contenedores para la solucion
t = np.arange( a, b+h, h )
w = np.zeros((N+1,))

 # Colocamos los valores iniciales                      
t[0], w[0] = ValInicial 

# Aplicamos el metodo de Euler Modificado en el punto medio
for i in range(1,N+1): 
    w[i] = w[i-1] + h * f( t[i-1] + h/2.0, w[i-1] + h * f( t[i-1], w[i-1] ) / 2.0 )

# Recuperamos la solucion exacta
solExacta = raw_input("\nIngrese la Funcion de Solucion Exacta: y(t) = ")

def y( t ):
    return eval(solExacta)        

print("\n-----------------------------------------------------\n")
print("Valor de Euler Modificado en el Punto medio: \n")
print w[i]
print("\n-----------------------------------------------------\n")


error_abs = lambda y, w: np.sum( np.abs( y - w ) )
error_abs( y(t), w )

print("\n-----------------------------------------------------\n")
print("El Error en el metodo es: \n")
print (error_abs(y(t), w))
print("\n-----------------------------------------------------\n")

plot( t, w, 'r', label='Aproximacion' )
plot( t, y(t), 'g', label='Valor Exacto' )
title( "Modified Euler's Method, N="+str(N) )
xlabel('t') 
ylabel('y(t)')
legend(loc=4)
grid()
```

<center><h3>Método de Euler Heun</h3></center>


Partimos de la siguiente expresión:

\begin{align}
    x(t_{1}) = x(t_{0}) \int_{t_{0}}^{t_{1}} f(s, x(s)) ds
\end{align}

Si aplicamos ahora la fórmula del trapecio a la integral, tenemos que:

\begin{align}
    x(t_{1}) = x_{t_{0}} + \frac{t_{1} - t_{0}}{2} \cdot [f \cdot (t_{0}, x(t_{0})) + f \cdot (t_{1}, x(t_{1})) ]
\end{align}

O sea:

\begin{align}
    x_{1} = x_{0} + \frac{h}{2} [ f \cdot (t_{0}, x_{0}) + f \cdot (t_{1}, x_{1})].
\end{align}

Haciendo lo mismo en los puntos sucesivos, la recurrencia asociada al método de Heun es:

\begin{align}
    x_{k+1} = x_{k} + \frac{h}{2} [ f \cdot (t_{k}, x_{k}) + f \cdot (t_{k+1}, x_{k+1})], k = 0, 1, ..., n-1 
\end{align}

Se dice que el método de Heun es un método implícito. Esto significa que, en cada paso $ x_{k+1} $ está definido en función de sí mismo. Por tanto, salvo que $f$ sea lineal en $x$, necesitamos resolver una ecuación no lineal. 

Para solventar este inconveniente una posibilidad es calcular un valor previo $x^{*}_{k+1}$ y, a partir, de que el determina $x_{k+1}$. Tenemos así que Heun es un ejemplo de los llamados métodos multi-paso predictor-corrector. 

Por lo que, el método de Heun viene dado porla recurrencia:

\begin{align}
    Predicción: x^{*}_{k+1} = x_{k} + h \cdot f(t_{k}, x_{k}) 
    \\\
    Correción: x_{k+1} = x_{k} + \frac{h}{2} \cdot [ f \cdot (t_{k}, x_{k}) + f \cdot ( t_{k+1}, x^{*}_{k+1} )  ] 
\end{align}

para $k = 0, 1, . . . , n − 1$. En este método, las predicciones se calculan aplicando el método de Euler.

En Python, la secuencia es la siguiente:


```python
for i in range(1,N+1):              # aplicamos el Heun´s Euler Method
        f1 = f( t[i-1], w[i-1] )
        f2 = f( t[i], w[i-1] + h * f1 )
        w[i] = w[i-1] + h * ( f1 + f2 ) / 2.0
```

\end{align}

donde:

\begin{align}
    y_{i+1}^{0} = y_{i} + f \cdot (x_{i}, y_{i}) \cdot h
\end{align}
